```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
```

```python
def get_tweets(query):
    
    tweets_list = []
    
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
        tweets_list.append([tweet.date, 
                            tweet.id, 
                            tweet.content, 
                            tweet.user.username, 
                            tweet.replyCount, 
                            tweet.retweetCount, 
                            tweet.likeCount, 
                            tweet.quoteCount])
        
    dft = pd.DataFrame(tweets_list, columns=['Datetime', 
                                             'Tweet_Id', 
                                             'Text', 
                                             'Username', 
                                             'Reply_Count', 
                                             'Retweet_Count', 
                                             'Like_Count', 
                                             'Quote_Count'])
    
    dft['Date'] = dft.Datetime.apply(lambda x: x.date())
    
    return dft
```

```python
# Scrape
query = 'lang:en AND min_faves:100 AND (#bitcoin OR bitcoin OR btc OR #btc) since:2020-04-01 until:2021-03-15'
df_tw = get_tweets(query)

# Drop duplicates
df_tw.drop_duplicates(subset = 'Text',keep = 'first', inplace = True)

# Store the row dataframe
df_tw.to_csv('df_tw_row.csv', index=False)

# View the dataframe
df_tw
```
