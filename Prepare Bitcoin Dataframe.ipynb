```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# Read the btc csv
df_btc = pd.read_csv('BTC-USD.csv')
df_btc.Date = pd.to_datetime(df_btc.Date)

# Create the analysis df and compute derived btc variables
#df_btc = btc_df[(df_btc.Date >= pd.to_datetime('20-07-2020', format='%d-%m-%Y')) & \
#       (df_btc.Date < pd.to_datetime('15-03-2021', format='%d-%m-%Y'))]
df_btc = df_btc[['Date', 'Close']]
df_btc = df_btc.dropna()

# Compute BTC Close price for next 1/2/3 days
df_btc['Close_next1'] = df_btc.Close.shift(-1).to_list()
df_btc['Close_next2'] = df_btc.Close.shift(-2).to_list()
df_btc['Close_next3'] = df_btc.Close.shift(-3).to_list()

# Compute the difference between the BTC Close price in 1/2/3 days wrt today
df_btc['Close_diff_next1'] = (df_btc.Close.shift(-1) - df_btc.Close).to_list()
df_btc['Close_diff_next2'] = (df_btc.Close.shift(-2) - df_btc.Close).to_list()
df_btc['Close_diff_next3'] = (df_btc.Close.shift(-3) - df_btc.Close).to_list()

# Compute the above differences relative to the BTC Clsoe price today 
df_btc['Close_diff_next1_rel'] = (1. + df_btc.Close_diff_next1/df_btc.Close).to_list()
df_btc['Close_diff_next2_rel'] = (1. + df_btc.Close_diff_next2/df_btc.Close).to_list()
df_btc['Close_diff_next3_rel'] = (1. + df_btc.Close_diff_next3/df_btc.Close).to_list()

# Compute the sign of the differences above
df_btc['Close_diff_next1_sgn'] = np.sign(df_btc.Close_diff_next1).to_list()
df_btc['Close_diff_next2_sgn'] = np.sign(df_btc.Close_diff_next2).to_list()
df_btc['Close_diff_next3_sgn'] = np.sign(df_btc.Close_diff_next3).to_list()

# Compute rolling average and spread of the BTC Close price
df_btc['Close_last7_mean'] = df_btc.Close.rolling(7).mean().to_list()
df_btc['Close_last5_mean'] = df_btc.Close.rolling(5).mean().to_list()
df_btc['Close_last3_mean'] = df_btc.Close.rolling(3).mean().to_list()
df_btc['Close_last2_mean'] = df_btc.Close.rolling(2).mean().to_list()
df_btc['Close_last7_std'] = df_btc.Close.rolling(7).std().to_list()
df_btc['Close_last5_std'] = df_btc.Close.rolling(5).std().to_list()
df_btc['Close_last3_std'] = df_btc.Close.rolling(3).std().to_list()
df_btc['Close_last2_std'] = df_btc.Close.rolling(2).std().to_list()

# Compute the absolute and relative diff of the BTC Close price today wrt last 1/2/3/5/7 days
df_btc['Close_diff_last1'] = (df_btc.Close - df_btc.Close.shift(1)).to_list()
df_btc['Close_diff_last2'] = (df_btc.Close - df_btc.Close.shift(2)).to_list()
df_btc['Close_diff_last3'] = (df_btc.Close - df_btc.Close.shift(3)).to_list()
df_btc['Close_diff_last5'] = (df_btc.Close - df_btc.Close.shift(5)).to_list()
df_btc['Close_diff_last7'] = (df_btc.Close - df_btc.Close.shift(7)).to_list()
df_btc['Close_diff_last1_rel'] = ((df_btc.Close - df_btc.Close.shift(1))/df_btc.Close).to_list()
df_btc['Close_diff_last2_rel'] = ((df_btc.Close - df_btc.Close.shift(2))/df_btc.Close).to_list()
df_btc['Close_diff_last3_rel'] = ((df_btc.Close - df_btc.Close.shift(3))/df_btc.Close).to_list()
df_btc['Close_diff_last5_rel'] = ((df_btc.Close - df_btc.Close.shift(5))/df_btc.Close).to_list()
df_btc['Close_diff_last7_rel'] = ((df_btc.Close - df_btc.Close.shift(7))/df_btc.Close).to_list()

# Remove rows with NaN values
df_btc = df_btc.dropna()
```
